```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv("cosimo.csv")
```

```python
df.head()
```

```python
def parse_date(date_string):
    try:
        return datetime.strptime(date_string, "%d/%m/%Y").date()
    except ValueError as e:
        print(e)
        return np.nan
```

```python
df["parsed_date"] = df["Modern Date"].dropna().apply(parse_date)
```
